## Aula 3 - Histogramas

**Histograma** é um gráfico que permite visualizar a distribuição de intensidade de uma imagem, com os pixels variando entre 0 e 255 (geralmente).

É uma forma de entender a imagem, onde conseguimos ter intuições sobre o contraste, brilho e intensidade da imagem.

### Encontrando o histograma usando o OpenCV

Vamos importar as bibliotecas necessárias e carregar a imagem de teste:

```python
import cv2 # OpenCV
import matplotlib.pyplot as plt # Matplotlib
import numpy as np # Numpy
```

```python
img = plt.imread('lenna.jpg')
plt.imshow(img)
```

Usaremos a função [hist()](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) do Matplotlib. Essa função irá fazer o plot do histograma na escala de pixels especificada. Passaremos como argumento a resposta da função [ravel()](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.ravel.html), que é o vetor da imagem em 1-D.

```python
img_vector = img.ravel()
print(img_vector)
plt.hist(img_vector,256,[0,256])
plt.show()
```

Ou, o equivalente para OpenCV:

```python
h = cv2.calcHist([img.ravel()], [0], None, [256], [0,256])
plt.plot(h)
plt.show()
```

Vamos agora analisar a distribuição por canal. Usaremos apenas o OpenCV neste exemplo:

```python
color = ('b','g','r')
for i,col in enumerate(color):
    h = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(h, color = col)
    plt.xlim([0,256])
plt.show()
```
